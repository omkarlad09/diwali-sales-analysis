#eploratory data analysis

gender

```python
ax= sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

from above graph female do more purchase as compaire to male

age

```python
df.columns
```

```python
age_grp = sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in age_grp.containers:
    age_grp.bar_label(bars)
```

```python
sales_age_group= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age_group)
```

age grp from 26-35yr female spend more money 

```python
df.columns
```

```python
sales_state= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

most of order from uttar paradesh,mahashtra,karnata

```python
sales_state_purchase= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state_purchase)
```

marrital status

```python
marital_st=sns.countplot(data=df,x="Marital_Status",)
sns.set(rc={'figure.figsize':(5,5)})
for bars in marital_st.containers:
    marital_st.bar_label(bars)
```

```python
sale_marital= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=sale_marital,hue="Gender")
```

most amount spend by married women

Occupation

```python
sns.set(rc={"figure.figsize":(20,5)})
occuptaion=sns.countplot(data=df,x='Occupation')
for bars in occuptaion.containers:
    occuptaion.bar_label(bars)
```

```python
sales_by_occupation= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_by_occupation)
```

most of buyers from it,healtcare,avition

product catogory

```python
sns.set(rc={"figure.figsize":(25,5)})
product_catagory=sns.countplot(data=df,x='Product_Category')
for bars in product_catagory.containers:
    product_catagory.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_state)
```

most amount spend in food,clothing,electronis

product id

```python
sales_product= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_product)
```
